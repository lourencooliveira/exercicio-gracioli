1. Faça uma sub-rotina que receba um número inteiro e positivo N como parâmetro e retorne a soma dos números inteiros existentes entre o número 1 e N (inclusive).


```python
def soma_ate_n(n):
    # Verifica se o número informado é válido (inteiro positivo)
    if n <= 0:
        return 0  # Se N não for positivo, não há números para somar

    soma = 0        # Variável acumuladora, começa em 0
    contador = 1     # Contador que vai de 1 até N

    while contador <= n:    # Repete enquanto o contador não ultrapassar N
        soma = soma + contador   # Adiciona o valor atual do contador à soma
        contador = contador + 1  # Incrementa o contador em 1

    return soma     # Retorna o valor final da soma


# --- Testando a sub-rotina ---
n = int(input("Digite um número inteiro positivo: "))  # Lê o valor de N
resultado = soma_ate_n(n)        # Chama a sub-rotina passando N
print(f"A soma de 1 até {n} é: {resultado}")  # Exibe o resultado
```

3. Elabore uma sub-rotina que receba dois números como parâmetros e retorne 0, se o primeiro número for divisível pelo segundo número. Caso contrário, deverá retornar o próximo divisor.


```python
class CalculadoraDivisores:
    @staticmethod
    def eh_divisivel(dividendo: int, divisor: int) -> bool:
        if divisor == 0:
            return False
        return dividendo % divisor == 0

    @staticmethod
    def encontrar_proximo_divisor(numero: int, limite_inferior: int) -> int:
        proximo = limite_inferior + 1
        while proximo <= abs(numero):
            if numero % proximo == 0:
                return proximo
            proximo += 1
        return numero


class VerificadorDivisibilidade:
    def __init__(self, calculadora: CalculadoraDivisores):
        self._calculadora = calculadora

    def avaliar(self, num1: int, num2: int) -> int:
        if self._calculadora.eh_divisivel(num1, num2):
            return 0
        return self._calculadora.encontrar_proximo_divisor(num1, num2)
```

```python
# Função para ler os dados dos habitantes
def ler_dados():
    idades = []
    sexos = []
    salarios = []
    filhos = []

    for i in range(15):
        print(f"\nHabitante {i + 1}")

        idade = int(input("Idade: "))
        sexo = input("Sexo (M/F): ").upper()
        salario = float(input("Salário: R$ "))
        num_filhos = int(input("Número de filhos: "))

        idades.append(idade)
        sexos.append(sexo)
        salarios.append(salario)
        filhos.append(num_filhos)

    return idades, sexos, salarios, filhos


# Função para calcular a média dos salários
def media_salarios(salarios):
    return sum(salarios) / len(salarios)


# Função para encontrar a menor idade
def menor_idade(idades):
    return min(idades)


# Função para encontrar a maior idade
def maior_idade(idades):
    return max(idades)


# Função para contar mulheres com 3 filhos
# e salário até R$ 500,00
def mulheres_tres_filhos(sexos, salarios, filhos):
    contador = 0

    for i in range(len(sexos)):
        if sexos[i] == "F" and filhos[i] == 3 and salarios[i] <= 500:
            contador += 1

    return contador


# ---------------- Programa Principal ----------------

idades, sexos, salarios, filhos = ler_dados()

media = media_salarios(salarios)
menor = menor_idade(idades)
maior = maior_idade(idades)
quantidade = mulheres_tres_filhos(sexos, salarios, filhos)

print("\n========== RESULTADOS ==========")
print(f"Média dos salários: R$ {media:.2f}")
print(f"Menor idade: {menor} anos")
print(f"Maior idade: {maior} anos")
print(f"Quantidade de mulheres com 3 filhos e salário até R$ 500,00: {quantidade}")
```

```python
def separar_vetores(X):
    A = []  # Elementos maiores que zero
    B = []  # Elementos menores ou iguais a zero

    for numero in X:
        if numero > 0:
            A.append(numero)
        else:
            B.append(numero)

    return A, B


# Programa principal
X = []

print("Digite 30 números inteiros:")

for i in range(30):
    numero = int(input(f"Elemento {i + 1}: "))
    X.append(numero)

A, B = separar_vetores(X)

print("\nVetor X:")
print(X)

print("\nVetor A (maiores que zero):")
print(A)

print("\nVetor B (menores ou iguais a zero):")
print(B)
```

20. A prefeitura de uma cidade fez uma pesquisa entre seus habitantes, coletando dados sobre o salário e o número de filhos.
```python
def pesquisa_populacao():
    total_salario = 0
    total_filhos = 0
    maior_salario = 0
    abaixo_380 = 0
    total_pessoas = 0

    while True:
        salario = float(input("Digite o salário: R$ "))
        filhos = int(input("Digite o número de filhos: "))

        total_salario += salario
        total_filhos += filhos
        total_pessoas += 1

        if salario > maior_salario:
            maior_salario = salario

        if salario < 380:
            abaixo_380 += 1

        continuar = input("Deseja cadastrar outra pessoa? (S/N): ").strip().upper()
        if continuar == "N":
            break

    media_salario = total_salario / total_pessoas
    media_filhos = total_filhos / total_pessoas
    percentual_abaixo_380 = (abaixo_380 / total_pessoas) * 100

    return media_salario, media_filhos, maior_salario, percentual_abaixo_380


# Programa principal
media_sal, media_fil, maior_sal, perc = pesquisa_populacao()

print("\nRESULTADOS DA PESQUISA")
print(f"Média de salário: R$ {media_sal:.2f}")
print(f"Média de filhos: {media_fil:.2f}")
print(f"Maior salário: R$ {maior_sal:.2f}")
print(f"Percentual com salário inferior a R$ 380,00: {perc:.2f}%")
```

Na opção 7: o programa termina.

```python
# Dicionário de vendedores: {codigo: nome} — ex: {1: "João", 2: "Maria"}
vendedores = {}

# Matriz 4 vendedores x 12 meses; vendas[0][0] = vendedor 1 no mês 1
vendas = [[0.0] * 12 for _ in range(4)]


def cadastrar_vendedor():
    codigo = int(input("Código do vendedor (1-4): "))
    if not (1 <= codigo <= 4):
        print("Código inválido.")
        return
    if codigo in vendedores:
        print("Vendedor já cadastrado.")
        return
    vendedores[codigo] = input("Nome do vendedor: ")
    print("Vendedor cadastrado com sucesso!")


def cadastrar_venda():
    codigo = int(input("Código do vendedor: "))
    if codigo not in vendedores:
        print("Vendedor não encontrado.")
        return
    mes = int(input("Mês (1-12): "))
    if not (1 <= mes <= 12):
        print("Mês inválido.")
        return
    if vendas[codigo - 1][mes - 1] != 0:
        print("Já existe uma venda para esse vendedor nesse mês.")
        return
    # codigo-1 e mes-1 porque listas começam no índice 0
    vendas[codigo - 1][mes - 1] = float(input("Valor da venda: R$ "))
    print("Venda cadastrada com sucesso!")


def consultar_venda_mes():
    codigo = int(input("Código do vendedor: "))
    if codigo not in vendedores:
        print("Vendedor não encontrado.")
        return
    mes = int(input("Mês (1-12): "))
    if not (1 <= mes <= 12):
        print("Mês inválido.")
        return
    valor = vendas[codigo - 1][mes - 1]
    print(f"{vendedores[codigo]} - Mês {mes}: R$ {valor:.2f}")


def consultar_total_vendedor():
    codigo = int(input("Código do vendedor: "))
    if codigo not in vendedores:
        print("Vendedor não encontrado.")
        return
    # sum() soma todos os 12 meses automaticamente
    total = sum(vendas[codigo - 1])
    print(f"Total de vendas de {vendedores[codigo]}: R$ {total:.2f}")


def vendedor_mais_vendeu_mes():
    mes = int(input("Mês (1-12): "))
    if not (1 <= mes <= 12):
        print("Mês inválido.")
        return
    # Lista com o valor de cada vendedor naquele mês
    valores_do_mes = [vendas[i][mes - 1] for i in range(4)]
    maior = max(valores_do_mes)
    if maior == 0:
        print("Nenhuma venda registrada nesse mês.")
        return
    # index() retorna a posição (0-3); +1 converte para o código do vendedor (1-4)
    codigo_vencedor = valores_do_mes.index(maior) + 1
    nome = vendedores.get(codigo_vencedor, "Não cadastrado")
    print(f"Maior venda no mês {mes}: {nome} com R$ {maior:.2f}")


def mes_mais_vendas():
    # Soma as vendas dos 4 vendedores para cada um dos 12 meses
    totais = [sum(vendas[v][mes] for v in range(4)) for mes in range(12)]
    maior = max(totais)
    if maior == 0:
        print("Nenhuma venda registrada.")
        return
    # index() retorna posição (0-11); +1 converte para o número do mês (1-12)
    print(f"Mês com maior volume: {totais.index(maior) + 1} (R$ {maior:.2f})")


# Cada entrada liga o número digitado à descrição e à função correspondente
opcoes = {
    "1": ("Cadastrar vendedor",              cadastrar_vendedor),
    "2": ("Cadastrar venda",                 cadastrar_venda),
    "3": ("Consultar venda por mês",         consultar_venda_mes),
    "4": ("Total de vendas do vendedor",     consultar_total_vendedor),
    "5": ("Vendedor que mais vendeu no mês", vendedor_mais_vendeu_mes),
    "6": ("Mês com maior volume de vendas",  mes_mais_vendas),
    "7": ("Sair",                            None),
}

while True:
    print("\n=== CONTROLE DE VENDAS ===")
    for chave, (descricao, _) in opcoes.items():
        print(f"{chave}. {descricao}")

    escolha = input("Opção: ")
    if escolha not in opcoes:
        print("Opção inválida.")
        continue
    if escolha == "7":
        print("Encerrando...")
        break

    descricao, funcao = opcoes[escolha]
    print(f"\n--- {descricao.upper()} ---")
    funcao()
```

5. Faça um programa que leia o código, a descrição, o valor unitário e a quantidade em estoque dos 50 produtos comercializados por uma papelaria. Essas informações devem ser armazenadas em um registro do tipo vetor em ordem crescente de código. Depois da leitura, o programa deverá:

realizar uma rotina que mostre todos os produtos cuja descrição comece com determinada letra (informada pelo usuário);
mostrar todos os produtos com quantidade em estoque inferior a cinco unidades.

###roda para rodar o principal

```python
class Produto:
    def __init__(self, codigo, descricao, valor_unitario, qtd_estoque):
        self.codigo = codigo
        self.descricao = descricao
        self.valor_unitario = valor_unitario
        self.qtd_estoque = qtd_estoque

# --- Sub-rotinas do Sistema ---

def cadastrar_produtos():
    vetor_produtos = []
    # Para testes rápidos, você pode alterar o range de 50 para um número menor (ex: 3)
    print("--- CADASTRO DE 50 PRODUTOS ---")
    for i in range(3):
        print(f"\nProduto {i+1}:")
        codigo = int(input("Código: "))
        descricao = input("Descrição: ").strip()
        valor_unitario = float(input("Valor Unitário: R$ "))
        qtd_estoque = int(input("Quantidade em Estoque: "))

        novo_produto = Produto(codigo, descricao, valor_unitario, qtd_estoque)
        vetor_produtos.append(novo_produto)

    # Ordena o vetor em ordem crescente pelo atributo 'codigo'
    vetor_produtos.sort(key=lambda p: p.codigo)
    print("\nProdutos cadastrados e ordenados por código com sucesso!")
    return vetor_produtos


def alterar_produto(vetor_produtos):
    print("\n--- ALTERAR PRODUTO ---")
    codigo_busca = int(input("Digite o código do produto que deseja alterar: "))

    encontrado = False
    for prod in vetor_produtos:
        if prod.codigo == codigo_busca:
            print(f"Produto encontrado: {prod.descricao}")
            prod.descricao = input("Nova Descrição: ").strip()
            prod.valor_unitario = float(input("Novo Valor Unitário: R$ "))
            prod.qtd_estoque = int(input("Nova Quantidade em Estoque: "))
            print("Dados alterados com sucesso!")
            encontrado = True
            break

    if not encontrado:
        print("Produto não localizado pelo código informado.")


def buscar_por_letra(vetor_produtos):
    print("\n--- BUSCAR POR LETRA INICIAL ---")
    letra = input("Digite a letra inicial para buscar: ").strip().lower()

    print(f"\nProdutos que começam com '{letra.upper()}':")
    encontrou_algum = False
    for prod in vetor_produtos:
        if prod.descricao.lower().startswith(letra):
            print(f"Cód: {prod.codigo} | {prod.descricao} | R$ {prod.valor_unitario:.2f} | Estoque: {prod.qtd_estoque}")
            encontrou_algum = True

    if not encontrou_algum:
        print("Nenhum produto encontrado com essa inicial.")


def estoque_critico(vetor_produtos):
    print("\n--- PRODUTOS COM ESTOQUE BAIXO (INFERIOR A 5) ---")
    encontrou_algum = False
    for prod in vetor_produtos:
        if prod.qtd_estoque < 5:
            print(f"Cód: {prod.codigo} | {prod.descricao} | Estoque Atual: {prod.qtd_estoque}")
            encontrou_algum = True

    if not encontrou_algum:
        print("Nenhum produto com estoque crítico.")
```

###Executa para rodar o programa principal

```python
# --- Programa Principal ---

# 1. Leitura e armazenamento dos dados
produtos = cadastrar_produtos()

# 2. Menu Interativo para as rotinas solicitadas
while True:
    print("\n================ MENU ================")
    print("1. Alterar dados de um produto")
    print("2. Buscar produtos por letra inicial")
    print("3. Relatório de estoque baixo (< 5 unidades)")
    print("4. Sair")
    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        alterar_produto(produtos)
    elif opcao == "2":
        buscar_por_letra(produtos)
    elif opcao == "3":
        estoque_critico(produtos)
    elif opcao == "4":
        print("Encerrando o programa da papelaria. Até logo!")
        break
    else:
        print("Opção inválida! Tente novamente.")
```

7. Considere que exista um registro com os seguintes atributos: codigo_cliente e nome_cliente e outro registro com os seguintes atributos: N°_conta, valor_compra, codigo_cliente. Faça um programa que:

inclua clientes, não permitindo que dois clientes possuam o mesmo código;
inclua contas, verificando se o código do cliente informado já está cadastrado. Caso não esteja, não deverá permitir a inclusão;
remova determinado cliente. Antes de executar a remoção, deverá verificar se ele não está vinculado a alguma conta. Se possuir, deverá mostrar a mensagem Exclusão não permitida. Caso contrário, deverá proceder à exclusão.

```python
# Classe Cliente
class Cliente:
    def __init__(self, codigo_cliente, nome_cliente):
        self.codigo_cliente = codigo_cliente
        self.nome_cliente = nome_cliente


# Classe Conta
class Conta:
    def __init__(self, numero_conta, valor_compra, codigo_cliente):
        self.numero_conta = numero_conta
        self.valor_compra = valor_compra
        self.codigo_cliente = codigo_cliente


# Listas para armazenar os dados
clientes = []
contas = []


# Função para cadastrar cliente
def incluir_cliente():
    codigo = int(input("Código do cliente: "))

    # Verifica se o código já existe
    for cliente in clientes:
        if cliente.codigo_cliente == codigo:
            print("Erro! Já existe um cliente com esse código.")
            return

    nome = input("Nome do cliente: ")
    clientes.append(Cliente(codigo, nome))
    print("Cliente cadastrado com sucesso!")


# Função para cadastrar conta
def incluir_conta():
    codigo = int(input("Código do cliente: "))

    # Verifica se o cliente existe
    cliente_existe = False
    for cliente in clientes:
        if cliente.codigo_cliente == codigo:
            cliente_existe = True
            break

    if not cliente_existe:
        print("Cliente não cadastrado. Inclusão não permitida.")
        return

    numero = int(input("Número da conta: "))
    valor = float(input("Valor da compra: "))

    contas.append(Conta(numero, valor, codigo))
    print("Conta cadastrada com sucesso!")


# Função para remover cliente
def remover_cliente():
    codigo = int(input("Código do cliente para remover: "))

    # Verifica se existe conta vinculada
    for conta in contas:
        if conta.codigo_cliente == codigo:
            print("Exclusão não permitida.")
            return

    # Remove o cliente
    for cliente in clientes:
        if cliente.codigo_cliente == codigo:
            clientes.remove(cliente)
            print("Cliente removido com sucesso!")
            return

    print("Cliente não encontrado.")


# Função para listar clientes
def listar_clientes():
    print("\n--- CLIENTES ---")
    if len(clientes) == 0:
        print("Nenhum cliente cadastrado.")
    else:
        for cliente in clientes:
            print(f"Código: {cliente.codigo_cliente} | Nome: {cliente.nome_cliente}")


# Função para listar contas
def listar_contas():
    print("\n--- CONTAS ---")
    if len(contas) == 0:
        print("Nenhuma conta cadastrada.")
    else:
        for conta in contas:
            print(f"Conta: {conta.numero_conta} | Valor: R$ {conta.valor_compra:.2f} | Cliente: {conta.codigo_cliente}")


# Menu principal
while True:
    print("\n===== MENU =====")
    print("1 - Incluir Cliente")
    print("2 - Incluir Conta")
    print("3 - Remover Cliente")
    print("4 - Listar Clientes")
    print("5 - Listar Contas")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        incluir_cliente()
    elif opcao == "2":
        incluir_conta()
    elif opcao == "3":
        remover_cliente()
    elif opcao == "4":
        listar_clientes()
    elif opcao == "5":
        listar_contas()
    elif opcao == "0":
        print("Programa encerrado.")
        break
    else:
        print("Opção inválida!")
```

8. Foi feita uma estatística em 15 estados brasileiros para coletar dados sobre acidentes de trânsito. Em cada estado foram observados os seguintes aspectos:

nome do estado;
número de veículos que circularam nesse estado (em 2012);
número de acidentes de trânsito (em 2012).

Crie um programa que permita conhecer:

o maior e o menor índice de acidentes de trânsito e o nome dos estados em que eles ocorreram;
o percentual de veículos em cada estado;
a média de acidentes em cada um dos estados.

```python
class EstadoDados:
    def __init__(self, nome: str, veiculos: int, acidentes: int):
        self.nome = nome
        self.veiculos = veiculos
        self.acidentes = acidentes
        self.indice_acidentes = acidentes / veiculos if veiculos > 0 else 0.0


class EstatisticaTrânsito:
    def __init__(self):
        self.estados = []

    def adicionar_estado(self, estado: EstadoDados):
        self.estados.append(estado)

    def calcular_total_veiculos(self) -> int:
        total = 0
        for estado in self.estados:
            total += estado.veiculos
        return total

    def obter_maior_indice(self) -> EstadoDados:
        if not self.estados:
            return None
        maior = self.estados[0]
        for estado in self.estados:
            if estado.indice_acidentes > maior.indice_acidentes:
                maior = estado
        return maior

    def obter_menor_indice(self) -> EstadoDados:
        if not self.estados:
            return None
        menor = self.estados[0]
        for estado in self.estados:
            if estado.indice_acidentes < menor.indice_acidentes:
                menor = estado
        return menor

    def calcular_percentual_veiculo(self, estado: EstadoDados) -> float:
        total = self.calcular_total_veiculos()
        if total == 0:
            return 0.0
        return (estado.veiculos / total) * 100


class InterfaceUsuario:
    def __init__(self):
        self.processador = EstatisticaTrânsito()

    def executar(self):
        print("--- Entrada de Dados dos 15 Estados ---")
        for i in range(1, 16):
            print(f"\nDados do {i}º Estado:")
            nome = input("Nome do estado: ")
            veiculos = int(input("Número de veículos: "))
            acidentes = int(input("Número de acidentes: "))

            estado = EstadoDados(nome, veiculos, acidentes)
            self.processador.adicionar_estado(estado)

        self.exibir_relatorio()

    def exibir_relatorio(self):
        maior = self.processador.obter_maior_indice()
        menor = self.processador.obter_menor_indice()

        print("\n" + "="*50)
        print("              RELATÓRIO ESTATÍSTICO  ")
        print("="*50)

        if maior and menor:
            print(f"-> Maior índice de acidentes: {maior.indice_acidentes:.4f} em {maior.nome}")
            print(f"-> Menor índice de acidentes: {menor.indice_acidentes:.4f} em {menor.nome}\n")

        print("-" * 50)
        print(f"{'Estado':<15} | {'% Veículos':<12} | {'Média Acidentes':<15}")
        print("-" * 50)

        for estado in self.processador.estados:
            percentual = self.processador.calcular_percentual_veiculo(estado)
            print(f"{estado.nome:<15} | {percentual:>10.2f}% | {estado.acidentes:>15}")

        print("="*50)


if __name__ == "__main__":
    app = InterfaceUsuario()
    app.executar()
```

```python
documentos = []

total_receber = 0
total_juros = 0

for i in range(15):
    print(f"\nDocumento {i + 1}")

    numero = input("Número do documento: ")
    codigo = input("Código do cliente: ")
    vencimento = input("Data de vencimento: ")
    pagamento = input("Data de pagamento: ")
    valor = float(input("Valor da conta: "))

    dias_atraso = int(input("Dias de atraso (0 se não houve): "))

    juros = 0

    if dias_atraso > 0:
        juros = valor * 0.0002 * dias_atraso

    total_receber = total_receber + valor + juros
    total_juros = total_juros + juros

    documento = {
        "numero": numero,
        "codigo": codigo,
        "vencimento": vencimento,
        "pagamento": pagamento,
        "valor": valor,
        "juros": juros
    }

    documentos.append(documento)

print("\n--- DOCUMENTOS CADASTRADOS ---")

for doc in documentos:
    print("\nNúmero:", doc["numero"])
    print("Código do Cliente:", doc["codigo"])
    print("Vencimento:", doc["vencimento"])
    print("Pagamento:", doc["pagamento"])
    print("Valor da Conta: R$", doc["valor"])
    print("Juros: R$", round(doc["juros"], 2))

media_juros = total_juros / 15

print("\n--- RESULTADOS FINAIS ---")
print("Total geral a receber: R$", round(total_receber, 2))
print("Média dos juros: R$", round(media_juros, 2))
```

```python
# Listas
pacientes = []
medicos = []
consultas = []

dias_validos = ["segunda", "terca", "quarta", "quinta", "sexta"]


# a) Cadastro de pacientes
def cadastrar_paciente():
    cod = int(input("Código do paciente: "))

    for paciente in pacientes:
        if paciente["cod"] == cod:
            print("Erro! Já existe um paciente com esse código.")
            return

    nome = input("Nome: ")
    endereco = input("Endereço: ")
    fone = input("Telefone: ")

    pacientes.append({
        "cod": cod,
        "nome": nome,
        "endereco": endereco,
        "fone": fone
    })

    print("Paciente cadastrado com sucesso!")


# b) Cadastro de médicos
def cadastrar_medico():
    cod = int(input("Código do médico: "))

    for medico in medicos:
        if medico["cod"] == cod:
            print("Erro! Já existe um médico com esse código.")
            return

    nome = input("Nome: ")
    endereco = input("Endereço: ")
    fone = input("Telefone: ")

    medicos.append({
        "cod": cod,
        "nome": nome,
        "endereco": endereco,
        "fone": fone
    })

    print("Médico cadastrado com sucesso!")


# c) Cadastro de consultas
def cadastrar_consulta():
    numero = int(input("Número da consulta: "))
    dia = input("Dia da semana: ").lower()

    if dia not in dias_validos:
        print("Dia inválido!")
        return

    hora = input("Hora: ")
    cod_med = int(input("Código do médico: "))
    cod_pac = int(input("Código do paciente: "))

    # Verifica se médico existe
    medico_existe = False
    for medico in medicos:
        if medico["cod"] == cod_med:
            medico_existe = True

    if not medico_existe:
        print("Médico não encontrado.")
        return

    # Verifica se paciente existe
    paciente_existe = False
    for paciente in pacientes:
        if paciente["cod"] == cod_pac:
            paciente_existe = True

    if not paciente_existe:
        print("Paciente não encontrado.")
        return

    # Máximo de 2 consultas por médico por dia
    contador = 0
    for consulta in consultas:
        if consulta["cod_med"] == cod_med and consulta["dia"] == dia:
            contador += 1

    if contador >= 2:
        print("Esse médico já possui duas consultas nesse dia.")
        return

    consultas.append({
        "numero": numero,
        "dia": dia,
        "hora": hora,
        "cod_med": cod_med,
        "cod_pac": cod_pac
    })

    print("Consulta cadastrada com sucesso!")


# d) Pesquisa consultas de um médico em determinado dia
def pesquisar_consultas():
    cod = int(input("Código do médico: "))
    dia = input("Dia da semana: ").lower()

    print("\nConsultas encontradas:")

    encontrou = False

    for consulta in consultas:
        if consulta["cod_med"] == cod and consulta["dia"] == dia:
            print(consulta)
            encontrou = True

    if not encontrou:
        print("Nenhuma consulta encontrada.")


# e) Relatório de consultas de um dia
def relatorio_dia():
    dia = input("Dia da semana: ").lower()

    print(f"\nConsultas de {dia}")

    encontrou = False

    for consulta in consultas:
        if consulta["dia"] == dia:
            print(consulta)
            encontrou = True

    if not encontrou:
        print("Nenhuma consulta cadastrada para esse dia.")


# Menu principal
while True:
    print("\n=== CLÍNICA MÉDICA ===")
    print("1 - Cadastrar paciente")
    print("2 - Cadastrar médico")
    print("3 - Cadastrar consulta")
    print("4 - Pesquisar consultas de um médico")
    print("5 - Relatório de consultas por dia")
    print("0 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":
        cadastrar_paciente()

    elif opcao == "2":
        cadastrar_medico()

    elif opcao == "3":
        cadastrar_consulta()

    elif opcao == "4":
        pesquisar_consultas()

    elif opcao == "5":
        relatorio_dia()

    elif opcao == "0":
        print("Programa encerrado.")
        break

    else:
        print("Opção inválida!")
```

b) mostre todas as contas cadastradas;

d) mostre o somatório das contas de determinado cliente;

e) mostre todas as contas com saldo negativo;

f) mostre o ativo bancário (soma de todos os saldos).

```python
clientes = []
contas = []


def ler_inteiro(mensagem):
    while True:
        try:
            return int(input(mensagem))
        except ValueError:
            print("  Valor inválido! Por favor, digite apenas números.")


def ler_decimal(mensagem):
    while True:
        try:
            return float(input(mensagem))
        except ValueError:
            print("  Valor inválido! Por favor, digite um número (ex: 1500.00).")


def cadastrar_clientes():
    quantidade = ler_inteiro("Quantos clientes você quer cadastrar? ")
    for i in range(quantidade):
        print(f"\n--- Dados do {i + 1}º cliente ---")
        codigo = ler_inteiro("Código do cliente: ")
        nome = input("Nome: ")
        idade = ler_inteiro("Idade: ")
        endereco = input("Endereço: ")
        cpf_cnpj = input("CPF ou CNPJ: ")

        clientes.append({
            "codigo": codigo,
            "nome": nome,
            "idade": idade,
            "endereco": endereco,
            "cpf_cnpj": cpf_cnpj
        })

        cadastrar_contas(codigo, nome)


def cadastrar_contas(codigo_cliente, nome_cliente):
    qtd = ler_inteiro(f"Quantas contas o {nome_cliente} possui? (até 15) ")
    for j in range(qtd):
        print(f"  Conta {j + 1} de {nome_cliente}:")
        numero = ler_inteiro("  Número da conta: ")
        if numero == 0:
            print("  Número zero não é válido, essa conta foi ignorada.")
            continue
        saldo = ler_decimal("  Saldo atual: R$ ")
        contas.append({
            "numero": numero,
            "cliente": codigo_cliente,
            "saldo": saldo
        })


def mostrar_todas_as_contas():
    print("\n=== Todas as contas cadastradas ===")
    for conta in contas:
        print(f"Conta {conta['numero']} | Cliente {conta['cliente']} | Saldo: R$ {conta['saldo']:.2f}")


def mostrar_contas_do_cliente():
    codigo = ler_inteiro("\nDigite o código do cliente que você quer consultar: ")
    print(f"Contas encontradas para o cliente {codigo}:")
    encontrou = False
    for conta in contas:
        if conta["cliente"] == codigo:
            print(f"  Conta {conta['numero']} | Saldo: R$ {conta['saldo']:.2f}")
            encontrou = True
    if not encontrou:
        print("  Esse cliente não tem nenhuma conta cadastrada.")


def somatorio_do_cliente():
    codigo = ler_inteiro("\nDigite o código do cliente para somar os saldos: ")
    total = 0
    for conta in contas:
        if conta["cliente"] == codigo:
            total += conta["saldo"]
    print(f"O cliente {codigo} tem ao todo R$ {total:.2f} somando todas as suas contas.")


def mostrar_saldos_negativos():
    print("\n=== Contas no vermelho (saldo negativo) ===")
    achou = False
    for conta in contas:
        if conta["saldo"] < 0:
            print(f"  Conta {conta['numero']} | Cliente {conta['cliente']} | Saldo: R$ {conta['saldo']:.2f}")
            achou = True
    if not achou:
        print("  Nenhuma conta está negativa. Que beleza!")


def mostrar_ativo_bancario():
    total = 0
    for conta in contas:
        total += conta["saldo"]
    print(f"\nO ativo bancário (a soma de tudo que o banco guarda) é de R$ {total:.2f}.")


cadastrar_clientes()
mostrar_todas_as_contas()
mostrar_contas_do_cliente()
somatorio_do_cliente()
mostrar_saldos_negativos()
mostrar_ativo_bancario()
```

```python
from datetime import datetime, date

#Criando o cliente
class Cliente():
  def __init__(self, nome: str, cpf: str, rg: str, endereco: str, data_primeira_compra: date):
    self.nome = nome
    self.cpf = cpf
    self.rg = rg
    self.data_primeira_compra = data_primeira_compra
    self.vip = False
    self.valor_total_gasto = 0.0

#Criando a loja com funções pedidas
class Loja():
  def __init__(self):
    self.lista_clientes = []

  def cadastrar(self):
    nome = input("Insíra o nome do cliente: ")
    cpf = input("Insíra o CPF do cliente: ")
    rg = input("Insíra o RG do cliente: ")
    data_str = input("Data (DD/MM/AAAA): ")
    data = datetime.strptime(data_str, "%d/%m/%Y").date()
    endereco = input("Insíra o endereço do cliente")

    #Salvando o cliente na loja
    cliente = Cliente(nome, cpf, rg, endereco, data)
    self.lista_clientes.append(cliente)

  def atualizar_gasto(self, rg: str):
    #Buscando cliente na lista de clientes
    for cliente in self.lista_clientes:
      if cliente.rg == rg: #Se encontrado adiciona o valor ao seu total gasto
        valor = float(input("Insíra o valor gasto pelo cliente"))
        cliente.valor_total_gasto += valor
        if cliente.vip == False and cliente.valor_total_gasto >= 5000:
          cliente.vip = True
        break
    else:
        print("Cliente não encontrado.")

  def verificar_prazo(self):
    for cliente in self.lista_clientes:
      if cliente.vip == False:
        diferenca = date.today() - cliente.data_primeira_compra
        diferenca_numero = diferenca.days
        if diferenca_numero >= 180:
          cliente.valor_total_gasto = 0
```

19. Uma empresa de eletrodomésticos está realizando um sorteio de uma Ferrari F-50, do qual participarão todos os que comprarem pelo menos cinco produtos de uma só vez nas lojas autorizadas. Faça um programa que:

a) leia os dados de todos os clientes, como nome, data de nascimento, CPF, RG, cidade em que mora, endereço e quantidade de eletrodomésticos adquiridos pelo cliente. Deve-se incluir um campo para o número do registro, que vai de 1 até 9999;

b) faça o sorteio entre os clientes participantes. Somente os clientes que compraram mais que cinco equipamentos deverão participar. Será sorteado um número, e o cliente que tiver o registro com esse número será o ganhador.

```python
import random

clientes = []

qtd_clientes = int(input("Quantos clientes serão cadastrados? "))

for i in range(qtd_clientes):
    print(f"\n--- Cadastro do cliente {i+1} ---")

    while True:
        registro = int(input("Número do registro (1 a 9999): "))
        if 1 <= registro <= 9999:
            break
        else:
            print("Registro inválido! Digite um número entre 1 e 9999.")

    nome = input("Nome: ")
    data_nascimento = input("Data de nascimento: ")
    cpf = input("CPF: ")
    rg = input("RG: ")
    cidade = input("Cidade: ")
    endereco = input("Endereço: ")
    quantidade = int(input("Quantidade de eletrodomésticos comprados: "))

    cliente = {
        "registro": registro,
        "nome": nome,
        "data_nascimento": data_nascimento,
        "cpf": cpf,
        "rg": rg,
        "cidade": cidade,
        "endereco": endereco,
        "quantidade": quantidade
    }

    clientes.append(cliente)

registros_participantes = []

for cliente in clientes:
    if cliente["quantidade"] > 5:
        registros_participantes.append(cliente["registro"])

if len(registros_participantes) == 0:
    print("\nNenhum cliente está apto a participar do sorteio.")
else:
    numero_sorteado = random.choice(registros_participantes)

    vencedor = None
    for cliente in clientes:
        if cliente["registro"] == numero_sorteado:
            vencedor = cliente
            break

    print("\n===== RESULTADO DO SORTEIO =====")
    print(f"Número sorteado: {numero_sorteado}")
    print(f"Nome do vencedor: {vencedor['nome']}")
    print(f"Data de nascimento: {vencedor['data_nascimento']}")
    print(f"CPF: {vencedor['cpf']}")
    print(f"RG: {vencedor['rg']}")
    print(f"Cidade: {vencedor['cidade']}")
    print(f"Endereço: {vencedor['endereco']}")
    print(f"Quantidade comprada: {vencedor['quantidade']}")
```

20. Faça um programa contendo os serviços que uma oficina mecânica pode realizar: ordem de serviço (número da OS, data, valor, serviço realizado, cliente).
Leia as informações sobre várias ordens de serviço e determine, ao final, a média dos valores, o nome do cliente que realizou o serviço mais caro, junto com a descrição desse serviço e a data de sua realização.

```python
ordens = []

qtd_os = int(input("Quantas ordens de serviço deseja cadastrar? "))

for i in range(qtd_os):
    print(f"\n--- Cadastro da ordem de serviço {i+1} ---")

    numero = int(input("Número da OS: "))
    data = input("Data da realização: ")
    valor = float(input("Valor do serviço: R$ "))
    servico = input("Descrição do serviço realizado: ")
    cliente = input("Nome do cliente: ")

    ordem = {
        "numero": numero,
        "data": data,
        "valor": valor,
        "servico": servico,
        "cliente": cliente
    }

    ordens.append(ordem)

# Cálculo da média dos valores

soma_valores = 0

for ordem in ordens:
    soma_valores += ordem["valor"]

media = soma_valores / qtd_os

# Identificação da ordem de serviço mais cara

mais_cara = ordens[0]

for ordem in ordens:
    if ordem["valor"] > mais_cara["valor"]:
        mais_cara = ordem

print("\n========== RESULTADO ==========")
print(f"Média dos valores das ordens: R$ {media:.2f}")
print(f"Cliente do serviço mais caro: {mais_cara['cliente']}")
print(f"Serviço realizado: {mais_cara['servico']}")
print(f"Data da realização: {mais_cara['data']}")
print(f"Valor do serviço mais caro: R$ {mais_cara['valor']:.2f}")
```
